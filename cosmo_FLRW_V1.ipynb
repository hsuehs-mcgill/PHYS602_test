# The Friedmann Lemaitre Robertson Walker  cosmology

We will use the <code>LambdaCDM</code> package from <code>astropy.cosmology</code>

```python
import numpy as np
from astropy.cosmology import LambdaCDM

from astropy import constants as cst
from astropy import units as u
import matplotlib.pyplot as plt

%pylab inline
```

```python
cosmo0 = LambdaCDM(H0=70,Om0=0,Ode0=0)
cosmo1 = LambdaCDM(H0=70,Om0=0.5,Ode0=0)
cosmo2 = LambdaCDM(H0=70,Om0=1,Ode0=0)
```

### Some quantities today

```python
cosmo0.H0
```

```python
(1/cosmo0.H0.cgs).to(u.year)
```

```python
cosmo0.critical_density0
```

```python
cosmo0.hubble_distance
```

### Some quantities as function of redshift z

```python
z = np.logspace(-4,1,100)
```

```python
plt.plot(z,cosmo0.H(z))
plt.plot(z,cosmo1.H(z))
plt.plot(z,cosmo2.H(z))
plt.xlabel('z')
plt.ylabel(cosmo0.H(z).unit)
```

```python
plt.plot(z,cosmo0.comoving_distance(z))
plt.plot(z,cosmo1.comoving_distance(z))
plt.plot(z,cosmo2.comoving_distance(z))
plt.xlabel('z')
plt.ylabel(cosmo0.comoving_distance(z).unit)
```

```python
plt.plot(z,cosmo0.critical_density(z))
plt.plot(z,cosmo1.critical_density(z))
plt.plot(z,cosmo2.critical_density(z))
plt.xlabel('z')
plt.ylabel(cosmo0.critical_density(z).unit)
```

```python
plt.plot(z,cosmo0.age(z))
plt.plot(z,cosmo1.age(z))
plt.plot(z,cosmo2.age(z))
plt.xlabel('z')
plt.ylabel(cosmo0.age(z).unit)
```
